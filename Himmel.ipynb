```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pycbc.coordinates as co
from mpl_toolkits.mplot3d import Axes3D
from pycbc.distributions import sky_location
```

```python
#uniform_sky_distribution = sky_location.UniformSky(polar_bounds = (-0.5,0.5), azimuthal_bounds = (0,2.))
uniform_sky_distribution = sky_location.UniformSky()

solid_angle_samples = uniform_sky_distribution.rvs(size=10000)
spin_mag = np.ndarray(shape=(10000),dtype=float)
for i in range (0,10000):
    spin_mag[i]=1.
    
spinx, spiny, spinz = co.spherical_to_cartesian(spin_mag, solid_angle_samples['dec'], solid_angle_samples['ra'])

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(spinx, spiny, spinz, s=1)

ax.set_xlabel('Spin X Axis')
ax.set_ylabel('Spin Y Axis')
ax.set_zlabel('Spin Z Axis')
plt.show()
```
